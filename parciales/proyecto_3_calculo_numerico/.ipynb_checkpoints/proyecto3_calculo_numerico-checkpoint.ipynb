```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

def trazador_cubico(xi, yi):
    n = len(xi)
    
    # Valores h
    h = np.zeros(n-1, dtype=float)
    for j in range(0, n-1):
        h[j] = xi[j+1] - xi[j]
    
    # Sistema de ecuaciones
    A = np.zeros(shape=(n, n), dtype=float)
    B = np.zeros(n, dtype=float)
    
    A[0, 0] = 1
    A[n-1, n-1] = 1
    
    for i in range(1, n-1):
        A[i, i-1] = h[i-1]
        A[i, i] = 2 * (h[i-1] + h[i])
        A[i, i+1] = h[i]
        B[i] = 6 * ((yi[i+1] - yi[i]) / h[i] - (yi[i] - yi[i-1]) / h[i-1])
    
    # Resolver sistema de ecuaciones S
    S = np.linalg.solve(A, B)
    
    # Coeficientes
    a = np.zeros(n-1, dtype=float)
    b = np.zeros(n-1, dtype=float)
    c = np.zeros(n-1, dtype=float)
    d = np.zeros(n-1, dtype=float)
    
    for j in range(0, n-1):
        a[j] = (S[j+1] - S[j]) / (6 * h[j])
        b[j] = S[j] / 2
        c[j] = (yi[j+1] - yi[j]) / h[j] - (2 * h[j] * S[j] + h[j] * S[j+1]) / 6
        d[j] = yi[j]
    
    # Polinomios por tramos
    x = sym.Symbol('x')
    px_tabla = []
    
    for j in range(0, n-1):
        pxtramo = a[j] * (x - xi[j])**3 + b[j] * (x - xi[j])**2 + c[j] * (x - xi[j]) + d[j]
        pxtramo = pxtramo.expand()
        px_tabla.append(pxtramo)
        
        # Imprimir polinomio por tramo
        print(f'Polinomio en el intervalo [{xi[j]}, {xi[j+1]}]:')
        print(pxtramo)
        print()  # lnea en blanco para separar los polinomios
    
    return px_tabla

# Datos de prueba (x_data, y_data) para graficar la curva
x_data = np.array([0, 1, 2, 3, 4, 5])
y_data = np.array([0.5, 0.8, 1.0, 0.9, 1.2, 0.7])

# Tabla de polinomios por tramos con el trazador cbico sujeto
px_tabla = trazador_cubico(x_data, y_data)

# Puntos para graficar cada tramo
muestras = 100
x_vals = np.linspace(0, 5, muestras)
y_vals = np.zeros_like(x_vals)

# Evaluar polinomio por tramos y construir la curva
tramo = 0
while tramo < len(px_tabla):
    pxt = sym.lambdify('x', px_tabla[tramo])
    mask = np.where((x_vals >= x_data[tramo]) & (x_vals <= x_data[tramo+1]))
    y_vals[mask] = pxt(x_vals[mask])
    tramo += 1

# Grfica
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, 'ro', label='Puntos de datos')
plt.plot(x_vals, y_vals, label='Trazador cbico sujeto', color='blue')
plt.title('Trazador Cbico Sujeto para la Trayectoria de la Montaa Rusa')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
def gauss_jordan(A):
    Ab = A
    n = len(Ab)

    for i in range(n):  
        Ab[i]/=Ab[i][i]
        for j in range(i+1, n):
            factor = Ab[j][i]
            Ab[j] -= factor * Ab[i]
        
    for i in range(n-1, -1, -1):
        for j in range(i-1, -1, -1):
            factor = Ab[j][i]
            Ab[j] -= factor * Ab[i]
           
    solucion=[]
    for i in range(0,len(Ab[0])-1):
        solucion.append(Ab[i][len(Ab[0])-1])
    r=0
    for i in solucion:
        print(f"a{r}={str(i)}\n")
        r+=1
    return solucion

def minimos_cuadrados(x, y, n):
    array_auxiliar = np.zeros(n+1, dtype=float)
    resultado = np.zeros((n, n+1), dtype=float)
    for i in range(n):
        for j in range(n+1):
            if i == 0:
                if j == 0:
                    array_auxiliar[j] = len(x)
                elif j == n:
                    array_auxiliar[j] = np.sum(y)
                else:
                    array_auxiliar[j] = np.sum(x**j)
            else:
                if j == n:
                    array_auxiliar[j] = np.sum((x**i) * y)
                else:
                    array_auxiliar[j] = np.sum(x**(i+j))
        
        resultado[i] = array_auxiliar.copy()
    
    return resultado 
            
xi = np.array([0, 1, 2, 3, 4])
yi = np.array([1.1, 3.5, 2.8, 4.2, 5.0])
# Valores para graficar la curva ajustada
datos_x = np.linspace(min(xi), max(xi), 100)
datos_y = np.zeros(len(datos_x))
x=sym.Symbol("x")
n=3
resolucion= gauss_jordan(minimos_cuadrados(xi,yi,n))
polinomio=0
for i in range(0,len(resolucion)):
    if i == 0:
        polinomio+=resolucion[i]
    else:
        polinomio+=resolucion[i]*(x**i)

for i in range(0,len(datos_y)):
    if isinstance(polinomio, np.float64): 
        datos_y[i]=polinomio
    else:
        datos_y[i]=polinomio.subs(x,datos_x[i])

#lineas de error
for i in range(0,n):
    y=np.min([yi[i],datos_y[i]])
    y2=np.max([yi[i],datos_y[i]])
    plt.vlines(xi[i],y,y2,color='green',linestyle='dotted')

plt.plot(xi,yi,'o',label="Datos experimentales")
plt.title("Analisis estructural por minimos cuadradados")
plt.xlabel("Tensiones")
plt.ylabel("Compresiones")
plt.plot(datos_x,datos_y,label="Curva ajustada")
plt.legend()
plt.show()
```
